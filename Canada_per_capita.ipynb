```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r"C:\Users\sarva\Downloads\canada_per_capita_income.csv")
```

```python
df
```

```python
df.columns =['year','income']
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year,df.income,color="b")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
reg.predict(pd.DataFrame([[2020]], columns=['year']))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year,df.income,color="r")
plt.plot(df.year,reg.predict(df[['year']]),color="b")
plt.show()
```
